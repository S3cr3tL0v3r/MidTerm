## Teilaufgabe 7: RandomWalk-Algorithmus implementieren! 

Mit den vorhandenen Funktionen sind die Grundlagen gelegt für ein erstes Lernverfahren - den RandomWalk-Algorithmus
* wird auch zufällige Irrfahrt, Zufallsweg, etc bezeichnet

Idee: 
* Für die aktuellen Werte $a$ und $b$ wird der MSE-Wert berechnet. 
* Die Werte von $a$ und $b$ werden variiert, indem jeweils ein zufälliger Wert einer Normalverteilung (Mittelwert = 0 und Standardabweichung = 1) zu den aktuellen Werten hinzugefügt wird. 
* für die neuen Werte $a$ und $b$ wird erneut ein MSE-Wert berechnet.
* Nun werden die beiden MSE-Werte verglichen: 
    * ist der neue Wert kleiner als der alte, 
        * hat der RandomWalk-Algorithmus einen besseres Modell gefunden
        * werden die neuen Werte verwendet und der Algorithmus erneut ausgeführt 
    * ist der neue Wert nicht kleiner als der alte
        * die alten Werte werden behalten und der Algorithmus erneut ausgeführt
<br>

```python
import numpy as np
import matplotlib.pyplot as plt
import random as ran

class RandomWalkDemo:
    def __init__(self):
        self.a = 50.0
        self.b = 0.0
        self.ranstate = 42
        self.rgen = ran.random()
        
    def plot(self):
        plt.title("Data")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.plot(x, y)
        
    def plotwithmodel(self):
        plt.title("Data")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.plot(x, y,'.k')
        xa = self.predict()
        plt.plot(x, xa)
        
    def predict(self):
        xp = np.array(x)
        xp = self.a + self.b * xp
        return xp
        
    def meansquareerror(self):
        return (1/y.size) * np.sum(np.square(self.predict() - y))
        
    def randomwalkfit(self,rounds):
        lista = []
        listb = []
        listmse = []
        self.plotwithmodel()
        for i in range(rounds + 1):
            alt_a = self.a
            lista += [self.a]
            alt_b = self.b
            listb += [self.b]
            alt_mse = self.meansquareerror()
            listmse += [alt_mse]
            self.a = self.a + np.random.randn()
            self.b = self.b + np.random.randn()
            if(alt_mse <= self.meansquareerror()):
                self.a = alt_a
                self.b = alt_b
        self.plotwithmodel()
        
        return lista,listb,listmse
    
b = RandomWalkDemo()
b.randomwalkfit(100000)
```
