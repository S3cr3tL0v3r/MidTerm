## Teilaufgabe 4: Phase 1: Exponentielles Wachstum - Doppelzeit

Das exponentielle Wachstum ist eine Eigenschaft der Ausbreitung von Epidemien. 
Mathematisch läßt sich diese Phase durch die folgende Funktion beschreiben: 

$$
N(t) = A \cdot e^{b \cdot t}
$$

In dieser Phase spielte die Verdoppelungszeit $T_D$ eine wesentliche Rolle. Das ist die Zeit, in der sich die Anzahl der Infektionen verdoppelt. Sie kann wie folgt aus der obigen Gleichung bestimmt werden.

$$
T_D = \frac{ln 2}{b}
$$

In der ersten Phase der Pandemie lag ein solches Wachstum vor. Am 18. März hat die Regierung den sogenannten Lockdown beschlossen. Bestimmen sie für das Land Baden-Württemberg die Verdoppelungszeit $T_D$ für den Zeitraum vom Beginn der Datenerfassung bis zum 18. März.2020 (einschliesslich). 

* Verwenden Sie die Funktion `curve_fit()` Bibliothek `scipy.optimize`
* Definieren Sie eine Funktion `myExpo(x, A, b)` für den exponentiellen Verlauf
* Definieren Sie eine Funktion `myDoppelzeit(b)` für die Berechnung der Verdoppelungszeit
* Zeichnen Sie ein Diagramm mit den Daten dieser Tage und mit der ermittelten Exponentialfunktion! 
    * verwenden Sie `matplotlib.pyplot`

```python
import scipy.optimize as scop
def myExpo(x,a,b):
    return a * np.exp(b * x)
def myDoppelzeit(b):
    return np.log(2) / 2
result, cov = scop.curve_fit(myExpo, range(21), data.loc['Summe'][:21])
plt.xlabel("Datum")
plt.ylabel("Infizierte")
plt.title("Infizierte in den ersten 16 Tagen")
plt.plot(data.columns[:21].values, data.loc['Summe'][:21],label="Daten")
plt.plot(data.columns[:21].values, myExpo(range(21), result[0],result[1]),label="Exponentialfunktion")
plt.legend()
print("Doppelzeit: "+str(myDoppelzeit(result[1])))
```
